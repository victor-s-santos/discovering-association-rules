## Transforming age
>How we have a large amount of single value of age, we need to make this in a interval

```python
print(df['idade'].unique())
print(len(df['idade'].unique()))
```

* __Age interval:__

>Age < 17

> 17 < Age < 22

> 22 < Age < 30

>30 < Age < 40

> Age > 40

```python
def age_interval(df_column):
    """Must return the respective age interval""" 
    if int(df_column) < 17:
        return("Age less than 17")
    elif 17 <= int(df_column) and int(df_column) < 22:
        return("Age between 17 and 22")
    elif 22 <= int(df_column) and int(df_column) < 30:
        return("Age between 22 and 30")
    elif 30 <= int(df_column) and int(df_column) < 40:
        return("Age between 30 and 40")
    elif int(df_column) > 40:
        return("Older than 40")
```

```python
df['age_interval'] = df['idade'].apply(age_interval)
df
```

## Total pontos (total points)
>The same problem as in age values: how I have a large variability between values, I need to use an interval.

>As in the previous section, the frequency distribution is in the respective notebook.

```python
len(df['total_pontos'].unique())
```

```python
def score_interval(df_column):
    """Must return the respective points interval""" 
    if float(df_column) < 9:
        return("Score > 9")
    elif 9 <= float(df_column) and float(df_column) < 18:
        return("Score 9 - 18")
    elif 18 <= float(df_column) and float(df_column) < 27:
        return("Score 18 - 27")
    elif 27 <= float(df_column) and float(df_column) < 36:
        return("Score 27 - 36")
    elif 36 <= float(df_column) and float(df_column) < 45:
        return("Score 36 - 45")
    elif 45 <= float(df_column) and float(df_column) < 54:
        return("Score 45 - 54")
    elif 54 <= float(df_column) and float(df_column) < 63:
        return("Score 54 - 63")
    elif 63 <= float(df_column) and float(df_column) < 72:
        return("Score 63 - 72")
    elif 72 <= float(df_column) and float(df_column) < 81:
        return("Score 72 - 81")
    elif float(df_column) >= 81:
        return("Score > 80")
```

```python
df['score_interval'] = df['total_pontos'].apply(score_interval)
df['score_interval']
```

## Course Class Period
>We are going to create a column with the class period by course. This information is known and given by the university previosly.

```python
def give_period(df_column):
    """Must returns respective class period by course"""
    integral = ['ENGENHARIA AGRONÔMICA', 'AGROPECUÁRIA', 'INFORMÁTICA', 'TÉCNICO EM AGROPECUÁRIA',
               'ENSINO MÉDIO E TÉCNICO EM INFORMÁTICA', 'ENSINO MÉDIO E TÉCNICO EM ALIMENTOS', 'ENSINO MÉDIO E TÉCNICO EM AGROPECUÁRIA']
    noturno = ['TÉCNICO EM INFORMÁTICA - NOTURNO', 'CIÊNCIA DA COMPUTAÇÃO - BACHARELADO', 'CIÊNCIAS BIOLÓGICAS - LICENCIATURA',
              'EDUCAÇÃO FÍSICA - BACHARELADO - NOTURNO', 'TÉCNICO EM MEIO AMBIENTE', 'TÉCNICO EM ENFERMAGEM - NOTURNO',
              'EDUCAÇÃO FÍSICA - LICENCIATURA - NOTURNO', 'TÉCNICO EM SEGURANÇA DO TRABALHO']
    vespertino = ['EDUCAÇÃO FÍSICA - LICENCIATURA - VESPERTINO', 'TÉCNICO EM INFORMÁTICA - VESPERTINO', 'EDUCAÇÃO FÍSICA - BACHARELADO - VESPERTINO']
    dicio = {'integral': integral, 'noturno': noturno, 'vespertino': vespertino}
    for key, value in dicio.items():
        if df_column in value:
            return(key)
```

## Course Class Period
>We are going to create a column with the class period by course. This information is known and given by the university previosly.

```python
df['class_period'] = df['curso'].apply(give_period)
df['class_period']
```

## Course Name
>We are going to redefine course name, removing its class period and bachelor type.

>This way: TÉCNICO EM INFORMÁTICA - NOTURNO => INFORMÁTICA

```python
def refine_course_name(df_column):
    """Must returns only course name"""
    if str(df_column) == 'TÉCNICO EM INFORMÁTICA - NOTURNO' or \
    str(df_column) == 'TÉCNICO EM INFORMÁTICA - VESPERTINO':
        return('INFORMÁTICA')
    elif str(df_column) == 'TÉCNICO EM AGROPECUÁRIA' or \
    str(df_column) == 'ENSINO MÉDIO E TÉCNICO EM AGROPECUÁRIA':
        return('AGROPECUÁRIA')
    elif str(df_column) == 'CIÊNCIA DA COMPUTAÇÃO - BACHARELADO':
        return('CIÊNCIA DA COMPUTAÇÃO')
    elif str(df_column) == 'CIÊNCIAS BIOLÓGICAS - LICENCIATURA':
        return('CIÊNCIAS BIOLÓGICAS')
    elif str(df_column) == 'EDUCAÇÃO FÍSICA - LICENCIATURA - VESPERTINO' or  \
    str(df_column) == 'EDUCAÇÃO FÍSICA - BACHARELADO - NOTURNO' \
    or str(df_column) == 'EDUCAÇÃO FÍSICA - LICENCIATURA - NOTURNO' \
    or str(df_column) == 'EDUCAÇÃO FÍSICA - BACHARELADO - VESPERTINO':
        return('EDUCAÇÃO FÍSICA')
    elif str(df_column) == 'TÉCNICO EM MEIO AMBIENTE':
        return('MEIO AMBIENTE')
    elif str(df_column)== 'TÉCNICO EM ENFERMAGEM - NOTURNO':
        return('ENFERMAGEM')
    elif str(df_column) == 'TÉCNICO EM SEGURANÇA DO TRABALHO':
        return('SEGURANÇA DO TRABALHO')
    else:
        return(df_column)
```

```python
df['course_name'] = df['curso'].apply(refine_course_name)
df['course_name']
```

```python
#Exporting data
try:
    df.to_csv('source/pre_processed1.csv', index=False)
    print('CSV file has been created!')
except:
    print('An exception occurred!')
```
