```python
import pandas as pd
import matplotlib.pyplot as plt
from numpy import genfromtxt
```

```python
#reading the csv
df_total_points = pd.read_csv('source/pre_processed0.csv')
df_total_points
```

```python
histogram = plt.hist(df_total_points['total_pontos'])
_ = plt.xlabel('Total Points')
_ = plt.ylabel('Occurrences')
```

```python
#Analysing the interval
for i in range(len(histogram[1])):
    print(histogram[1][i])
```

```python
#looking for the diference between a value(class) and the next one
for i in range(1, len(histogram[1])):
    print(histogram[1][i] - histogram[1][i-1])
```
