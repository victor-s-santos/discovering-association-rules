```python
import pandas as pd

data = pd.read_csv('static/real-case.csv')
data
```

* ## The above table is showing the transactions for a week at a market in French



```python
transactions = []
for i in range(0, len(data.columns)):
    transactions.append([str(data.values[i, j]) for j in range(0, 4)])
transactions    
```

```python
from apyori import apriori

rules = list(apriori(transactions, 
                     min_support = 0.2,
                     min_confidence = 0.8,
                     min_lift = 2,
                     min_lenght = 2))
rules
```

* __An observation before continue coding:__
>    To return a considerable number of elements, I needed to set a much smaller number to my_support and to my_confidence

```python
rules = list(apriori(transactions, 
                     min_support = 0.03,
                     min_confidence = 0.8,
                     min_lift = 2,
                     min_lenght = 2))
len(rules)
```

```python
rules = [list(elements) for elements in rules]
rules[0:6]
```

```python
formated_rules = []
for j in range(0, len(rules)):
    formated_rules.append([list(element) for element in rules[j][2]])
formated_rules
```

## How to define the min values to these params?

* __Example:__ 

    `I am looking for an item which is sold 3 times a day. In this case, how many times is this item sold in week?`

* __Therefore, for support:__

>3(times sold) * 7(days at week) = 21
>21 / 7501(number of transactions in the list) = 2,79 * 10^-3 or 0,00279

* __This idea suggests that the greater the number of records, the lesser the support and confidence__
