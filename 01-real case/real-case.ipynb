```python
data = pd.read_csv('static/real-case.csv', header = None)
transactions = []
for i in range(0, len(data)):
    transactions.append([str(data.values[i, j]) for j in range(0, len(data.columns))])
transactions[0:3]
```

* __An observation before continue coding:__
>    To return a considerable number of elements, I needed to set a much smaller number to my_support and to my_confidence

```python
from apyori import apriori
rules = apriori(transactions, 
                     min_support = 0.003,
                     min_confidence = 0.2,
                     min_lift = 3.0,
                     min_lenght = 2)
rules = list(rules)
len(rules)
```

```python
results = list(rules)
results[0:5]
```

```python
results2 = [list(x) for x in results]
results2[0:5]
```

```python
formatedresults = []
for j in range(0, len(results2)):
    formatedresults.append([list(x) for x in results2[j][2]])
formatedresults[0:5]
```

## How to define the min values to these params?

* __Example:__ 

    `I am looking for an item which is sold 3 times a day. In this case, how many times is this item sold in week?`

* __Therefore, for support:__

>3(times sold) * 7(days at week) = 21
>21 / 7501(number of transactions in the list) = 2,79 * 10^-3 or 0,00279

* __This idea suggests that the greater the number of records, the lesser the support and confidence__
